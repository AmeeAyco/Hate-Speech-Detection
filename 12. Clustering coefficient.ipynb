# Clustering coefficient

In graph theory, **a clustering coefficient is a measure of the degree to which nodes in a graph tend to cluster together.** Evidence suggests that in most real-world networks, and in particular social networks, nodes tend to create tightly knit groups characterized by a relatively high density of ties; this likelihood tends to be greater than the average probability of a tie randomly established between two nodes

## Types of Clustering coefficient

**There are two versions of this measure exist**:

1. **Global clustering coefficient**: The global version was designed to give an overall indication of the clustering in the network.

2. **Local clustering coefficient**: Local Clustering Coefficient of a node in a Graph gives an indication of the embeddedness of single nodes.

### Example 1

```python
# build graph
import networkx as nx
G=nx.erdos_renyi_graph(10,0.4)
nx.draw(G, node_size=1000, with_labels=True)
```

```python
# Output of Global CC 
cc=nx.average_clustering(G)
cc
```

```python
# Output of local CC
c=nx.clustering(G)
c
```

The above two values give us the global clustering coefficient of a network as well as local clustering coefficient of a network.

### Example 2

```python
import networkx as nx

G = nx.Graph()

G.add_edges_from([('A', 'B'), ('A', 'K'), ('B', 'K'), ('A', 'C'),
            ('B', 'C'), ('C', 'F'), ('F', 'G'), ('C', 'E'),
            ('E', 'F'), ('E', 'D'), ('E', 'H'), ('I', 'J')])

nx.draw(G, with_labels=True, node_size=1000)
```

```python
# returns a Dictionary with clustering value of each node
print(nx.clustering(G))
```

```python
# This returns clustering value of specified node
print(nx.clustering(G, 'C'))
```
